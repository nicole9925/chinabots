# Load Data

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
tweets = pd.read_csv("chinatweets1.csv")
```

```python
users1 = pd.read_csv("chinausers1.csv")
users2 = pd.read_csv("chinausers2.csv")
```

```python
users = pd.concat([users1, users2])
```

```python
grouped = pd.DataFrame(users.groupby("account_creation_date")["follower_count"].mean())
grouped = grouped.reset_index()
```

```python
users[["follower_count", "following_count"]].plot(kind="scatter", x="follower_count", y="following_count")
```
