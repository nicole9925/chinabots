```python
tweets1 = pd.read_csv("../data/tweets1.csv").drop_duplicates()
users1 = pd.read_csv("../data/users1.csv").drop_duplicates()

tweets2 = pd.read_csv("../data/tweets2.csv").drop_duplicates()
users2 = pd.read_csv("../data/users2.csv").drop_duplicates()

tweets3 = pd.read_csv("../data/tweets3_1.csv").drop_duplicates()
users3 = pd.read_csv("../data/users3.csv").drop_duplicates()

tweets4 = pd.read_csv("../data/tweets3_2.csv").drop_duplicates()
users4 = pd.read_csv("../data/users4.csv").drop_duplicates()

tweets5 = pd.read_csv("../data/tweets3_3.csv").drop_duplicates()

tweets6 = pd.read_csv("../data/tweets4.csv").drop_duplicates()
```

```python
tweet = pd.concat([tweets1,  tweets2, tweets3, tweets4, tweets5, tweets6])
user = pd.concat([users1, users2, users3, users4])
```

```python
english_df = tweet.loc[tweet["tweet_language"]=="en"].reset_index()
english_tweets = english_df["tweet_text"]

sentences = list(english_tweets.apply(lambda x:
                          re.sub("[\'\"\,\.\?\!\@\:\;|`\+\=\\|\\-\[\]\{\}\)\(\)]+", "", x)
                          .split()))
```

```python
from gensim.test.utils import common_texts
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(common_texts)]
```

```python
documents = [TaggedDocument(doc, [i]) for i, doc in enumerate(sentences)]
```

```python
model = Doc2Vec(documents, vector_size=5, window=2, min_count=1, workers=4)
```

```python
vecs = []
for document in sentences:
    try:
        vec= model[document]
        vecs.append(vec)
    except:
        print(document)
        vecs.append(None)
```

```python
sentences
```

```python
model.train(documents, total_examples=model.corpus_count, epochs=10, start_alpha=0.002, end_alpha=-0.016)
```

```python
kmeans_model = KMeans(n_clusters=4, init='k-means++', max_iter=100) 
```

```python
labels=kmeans_model.labels_.tolist()
```

```python
l = kmeans_model.fit_predict(model.docvecs.doctag_syn0)
```

```python
pca = PCA(n_components=2).fit(model.docvecs.doctag_syn0)
datapoint = pca.transform(model.docvecs.doctag_syn0)
```

```python
x = [i[0] for i in datapoint]
y = [i[1] for i in datapoint]
df = pd.DataFrame({"tweets": english_tweets, "x": x, "y": y, "group": l})
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure
label1 = ["#FFFF00", "#008000", "#0000FF", "#800080"]
color = [label1[i] for i in labels]
plt.scatter(datapoint[:, 0], datapoint[:, 1], c=color)
centroids = kmeans_model.cluster_centers_
centroidpoint = pca.transform(centroids)
plt.scatter(centroidpoint[:, 0], centroidpoint[:, 1], marker="^", s=150, c="#000000")
plt.show()
```
