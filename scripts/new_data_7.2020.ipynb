```python
users = pd.read_csv("../data/users_7.2020.csv")
users.shape
```

```python
tweets = pd.read_csv("../data/tweets_7.2020.csv", low_memory=False)
tweets['tweet_time'] = pd.to_datetime(tweets['tweet_time'])
tweets = tweets[['userid', 'tweet_text', 'tweet_time', 'is_retweet']]
tweets.shape
```

```python
tweets.head()
```

```python
# group users by highest tweet count
high_activity_users = tweets.groupby('userid').count()
high_activity_users = high_activity_users.sort_values(by = 'tweet_text', ascending = False).index
high_activity_users
```
